```python
from bs4 import BeautifulSoup as BS
import csv
import requests
```

```python
url = 'http://www.patheos.com/blogs'
response = requests.get(url)
```

```python
response.status_code
response.headers
response.content
```

```python
from bs4 import BeautifulSoup as BS
```

```python
soup = BS(response.content, 'html.parser')
soup.prettify
```

```python
blog_dict = {}
```

```python
for blog in soup.find_all('div', attrs={"class":"col-sm-6"})[0:20]:
#    blogname = blog.find('div', attrs={'class':'title'})
    blogurl = blog.find('a')
    blog_dict[blog.text] = blogurl['href']
    
#    print(blogname.text, end="")
#    print(blogurl.get('href'))
```

```python
blog_dict=dict(map(str.strip,x) for x in blog_dict.items())
blog_dict
```
